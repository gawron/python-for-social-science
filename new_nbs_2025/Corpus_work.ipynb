### LIX Readability

LIX (abbreviation of Swedish läsbarhetsindex, "readibility index") is a readability measure indicating the difficulty of reading a text developed by Swedish scholar Carl-Hugo Björnsson. It is defined as a sum of two numbers: the average sentence length and the percentage of words of more than six letters.

```python
import nltk
from nltk.corpus import gutenberg


def lix(fileid):
    A = len(gutenberg.words(fileids=fileid))
    B = len(gutenberg.sents(fileids=fileid))
    C = len([w for w in gutenberg.words(fileids=fileid) if len(w) >= 7])
    return (A/B) + ((C*100)/A)
```

```python
import time


num_trials = 100

times=[]
for i in range(num_trials):
    if i%10==0 and i >0:
        print(f"{i} trials done!")
    start = time.time()
    list0 = []
    for fileid in gutenberg.fileids():
        list0.append((lix(fileid),fileid))
        list0.sort(reverse = True)
    #print(*list, sep = "\n")
    end=time.time()
    times.append(end-start)

print(sum(times)/num_trials)
```
