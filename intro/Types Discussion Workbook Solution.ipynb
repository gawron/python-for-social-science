2. <font color="red">The emails of the students in a class.</font>

   **A dictionary with string keys (the student names) and string values (the phone numbers)**.

   For those who answered a list:  The answer is literally correct, but doesn't work for a very likely use case. A very likely use to which this data will be put is to answer questions like this:  "What is Tallal's email address?" A list of email strings doesn't allow you to answer that question in any obvious or computationally efficient way. You might amend your proposal to be a list of PAIRS (tuples of length 2) of strings, where the first member of each pair is a student's name and the second is their phone number.  That would work for the most likely use case. Of course you might answer that a list of email strings would work for mailing messages out to the students as a group, and that's true, but it's a very limited use case.  So the first question you should always think about is "How am I going to use this data?" And if you get explicit about answering that use case question, the answer to what type to use will usually be clear. The next question is:  Have I chosen the most computationally efficient data type for my use case? This is where the list-of-pairs answer has a problem.  If you are really interested in going from names to emails (Name$\Rightarrow$Email), then that's a key-value relationship and a dictionary would be far more efficient than a list of pairs.

8.  <font color="red">Same as (7), but this time we want to be able to do word counts and answer questions like "How many times does the word *general* appear in War and Peace?"</font>   Include with your answer to this question a brief description of how you will do word counting, which could be a code snippet.

    **A list of words.  Possibly with an an associated dictionary whose keys are the words (strings) and whose values are integers (the word counts).**

    Some have suggested that we could continue using a string and just call the  `str.count()` method when we need a word count.  So the following code snippet (more or less computes the frequency of the word *general* in *War and Peace**, assuming `WP` is a string containing the entire novel.

    ```python
    >>> WP.count("general")  #  Call .count() on the string WP
    570
    >>> WP.split().count("general")  #  Call .count() on the list (of words) WP.split()
    315
    ```
    
    The second line of code show the problem with computing word counts directly from strings.  The two numbers computed differ because the string WP contains substrings like:
    
    ```
    general
    generals
    generalization
    generalize
    generalized
    generalizing
    ```

    all of which will be counted by the first counting method, but not the second. The second method, while flawed, has some advantages, since *generalization** is clearly not the same word as *general**.  But it still has problems, since one of the elements of the split will be *general,* (*general* with the comma attached), which does not `==` *general*, and we probably do want to count *general,* as an occurrence of the word *general*.  But the problem here is not so much getting the count to be accurate (this is fixable) as it is efficiency. Suppose you want to look the count of the word *general* 100 different times in a large project. Using `.count()`, you have to go through all 500,000 words each time, finding each token of *general* in that large haystack of words (That's what the `.count()` method does).   Unless you stored the result somehow.  So you **definitely want** to precompute the dictionary if you will be looking word counts up often.

    When you think about how you store the result of the computation, you want a data type that let's you go from word to count (Word $\Rightarrow$ Count). That's a key-value relation, so you want a dictionary.

    As far as computing that dictionary goes, that wasn't part of the question but:  Iterate through all 500,00 words ONCE, updating the count of a word whenever you come across it, storing the results in a dictionary. 
 
    ```python
    count_dict = dict()
    for word in WP.split():
       if word in count_dict:
          count_dict[word] += 1
       else:
          count_dict[word] = 1
     ```
    
    What is **the difference between the tasks in questions (7) and (8)**?
  
    For this question, we claimed that your answer should not be the string again; rather it should be a list of words, possibly accompanied by a dictionary whose keys are word strings and whose values are word counts.

    Why is a string not the best answer for this question? The string implicitly contains the information you were supposed to think about (the word counts) but it is not the same as the word counts. Many kinds of information are implicit in the string, for example, the number of characters in it or the number of words or the number of occurrences of the letter "n" or the social network of the novel's characters based on the scenes they occur in together.  All are pieces of information that can be computed from the *War and Peace* string; they are all different from each other and from the string. And we can talk about how to store each of those kinds of information without worrying about how to compute them.
    
    So is the following true? Always precompute any needed information.  Always store it.  No.  Sometimes something is so easy to compute we don't want to waste space storing it.

    Summing up: For  (7) the correct answer is the string containing the entire content of the novel.  For (8) the answer is a list of words because we're interested in a word-level property.  And possibly dictionary (or a `Counter` from the `collections` module, which is a customized dictionary) if we're doing word counts often.
    
9.  <font color = "red">A partition of a set.</font>
    
    **a set of frozen sets** 
    
    Note that the obvious answer, a set of sets, won't work.  Let's try and store the **partition** P of 
    
    ```python
    {"if","the","link","an","park","ham"}
    ```
    
    by word length as 
    
     ```python
    P = {{"if", "an"}, {"the","ham"}, {"link", "park"}}
    ```
   
    This results in
    
    ```python
    ---------------------------------------------------------------------------
    TypeError                                 Traceback (most recent call last)
    Cell In[1], line 1
     ----> 1 {{"if", "an"}, {"the","ham"}, {"link", "park"}}
     
    TypeError: unhashable type: 'set'
    ```
    
    Oops!  We have violated the essential restriction on sets that their elements be immutable (= hashable!)
    What we need is
    
    ```python
    P = {frozenset({"if", "an"}), frozenset({"the", "ham"}), frozenset({"link", "park"})}
    ```
    
    Using a set of tuples also works:
    
    ```python
    P = {("an", "if"), ("ham", "the"), ("link", "park")}
    ```
    
    Where I've tried to forestall the complication of ordering information
    (where none was needed) by ordering the members of each pair alphabetically.  That way
    if I want to know whether the elements "the" and  "ham" fall in the same partition set,
    I put the two strings in alphabetical order and ask:
    
    ```python
    >>> ("ham", "the") in P
    ```

## Takeaways

1.  Look at how you **use** the information.   What kinds of questions will it be used to answer?  That will 
    generally take you right to the kind of data you need.
    
2.  For any container think about whether you've got the right kind of elements, with the kind of information you want to be computing on.  That's what we did in going from (7) to (8).  We went from a string to a list of words because we were interested in a word-level property, word counts.
    
3. Dictionaries can be thought as ways of storing the results of some computation on some set or some sequence S.  We **map** from S to f(S).  The word count example is informative.  We go from (a) a string S to (b) a sequence of words S' to (c) a mapping from the **vocabulary** of S' (set(S')) to word counts (Word_counts(set(S')).
