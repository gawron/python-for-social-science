```python
import numpy as npl
from matplotlib import pyplot as  plt

(f,ax) =plt.subplots(1,1)

v1 = np.array([2,10])
v2 = np.array([-1,2.5])

print(v1,type(v1))
print(v2,type(v2))

def plot_vec(vec,ax,c=None):
    x,y = vec
    ax.plot([0,x],[0,y],linestyle="--",c=c)


plot_vec(v1,ax)
plot_vec(v2,ax)
plot_vec(v1+v2,ax)

ax.axis('equal')

#ax.axhline(y=0, color='k')
#ax.axvline(x=0, color='k')
```

#  Multiplying a vector by scalar

```python

(f, ax) =plt.subplots(1,1)

v1 = np.array([2,10])
v2 = np.array([-1,2.5])

print("v1",v1)
print("-2 * v1", -2*v1)

plot_vec(v1,ax)
plot_vec(-2*v1,ax)
ax.axis('equal')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
```

## Applying  map to a vector

```python
(f, ax) =plt.subplots(1,1)


print("v1",v1)
print("Lv1", L@v1)

plot_vec(v1,ax)
plot_vec(L@v1,ax)
ax.axis('equal')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
```

```python
L = np.array([[2,1],[1,2]])
M = np.array([v1,v2]).T
points = L@M
(f, ax) =plt.subplots(1,1)


print("v1",v1)
print("Lv1", L@v1)

plot_vec(v1,ax,c="k")
plot_vec(v2,ax,"k")
plot_vec(points[:,0],ax,c="r")
plot_vec(points[:,1],ax,c="r")
ax.axis('equal')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
```
