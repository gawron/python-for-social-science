## Multivariate kde

#### Scaling and centering

```python
flip_mn = penguins.flipper_length_mm.mean()
flip_std = penguins.flipper_length_mm.std()
penguins["flip_sc_cen"] = (penguins.flipper_length_mm-flip_mn)/flip_std

bm_mn = penguins.body_mass_g.mean()
bm_std = penguins.body_mass_g.std()
penguins["bm_sc_cen"] = (penguins.body_mass_g-bm_mn)/bm_std
```

```python
fig, (ax0,ax1) = plt.subplots(1,2,figsize=(10,6.5))
sns.kdeplot(data=penguins, x="flipper_length_mm",y="body_mass_g", ax=ax0,bw_adjust=.9)

sns.kdeplot(data=penguins, x="flip_sc_cen",y="bm_sc_cen", ax=ax1,bw_adjust=1)
```

## Using colormaps in seaborn

A 2D histogram of the same data relation as the previous plot. Each square is a 2D bin.  The color encodes the count for the bin.  Thus the two contour centers from the previous plot have become  squares colored yellow-green and yellow (high-value colors) in this plot.

```python
# magma,viridis are  matplotlib colormaps; the seaborn colormaps illustated below dont work here.
sns.histplot(penguins, x="bill_depth_mm", y="bill_length_mm",cbar=True,cmap="magma")
```

You can also **create** an appropriate colormap from a `seaborn` palette and cast it in mpl form to provide the palette for the 2D histogram.

Note the controls allow you to tweak the values.  Try reversing for an effect  easy to see in the next plot:

```python
palette_ch = sns.choose_cubehelix_palette(as_cmap=None)
```

```python
mpl_ch = mpl.colors.ListedColormap(palette_ch)
sns.histplot(penguins, x="bill_depth_mm", y="bill_length_mm",cbar=True,cmap=mpl_ch)
```

Let's try a classic colormap application: a heat map.

Different "choosing" tools

```python
[nm for nm in dir(sns) if nm.startswith("choose")]
```

Since both high and negative and high positive values matter when taking correlations,
a diverging color map is appropriate.

```python
palette_diverging = sns.choose_diverging_palette(as_cmap=None)
```

```python
(_fig,_ax) = plt.subplots(1,1,figsize=(4,3))
penguins_numerical = penguins[["bill_length_mm", "bill_depth_mm", "body_mass_g","flipper_length_mm"]]

_ = sns.heatmap(penguins_numerical.corr(), cmap=palette_diverging)
```

Compare aesthetically  the native `pandas` rendering of the same correltion heat map, which includes
the correlation values:

```python
penguins_numerical.corr().style.background_gradient(cmap='coolwarm')
#penguins_numerical.corr().style.background_gradient(cmap=mpl_pd)
```

You can of course import the `seaborn` palette in a form which `pandas` can use, to show the
numbers and correlation intensities simultaneously while using a seaborn palette.

```python
mpl_pd = mpl.colors.ListedColormap(palette_diverging)
penguins_numerical.corr().style.background_gradient(cmap=mpl_pd)
```

## The seaborn graphic object interface

One of the pairwise plots from above redone to illustrate the object-oriented 
grammar_of_graphics interface to `seaborn`.  Color is species as above.  Also, point size is body mass.
The body mass binning has been done automatically.

```python
from seaborn import objects as so

(
    so.Plot(
        penguins, x="bill_length_mm", y="bill_depth_mm",
        color="species", pointsize="body_mass_g",
    )
    .add(so.Dot())
)
```

The simplified version. No graphic mappings.

```python
(
    so.Plot(penguins, x="bill_length_mm", y="bill_depth_mm")
    .add(so.Dot())
)
```

Even simpler. No graphic realizations chosen.

```python
(
    so.Plot(penguins, x="bill_length_mm", y="bill_depth_mm")
)
```
