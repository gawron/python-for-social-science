```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
female_data=(df[df['Gender']=='F'])
female_data.Year=female_data.Year//10*10
female_data.groupby(['Name','Year']).sum()                 
f1880=female_data[(female_data['Year']==1880)]
f1940=female_data[(female_data['Year']==1990)]
f1990=female_data[(female_data['Year']==1940)]
```

```python
f1880_name_set=set(f1880['Name'])
f1940_name_set=set(f1940['Name'])
f1990_name_set=set(f1990['Name'])
```

```python
print(len(f1880_name_set))
print(len(f1940_name_set))
print(len(f1990_name_set)) 
```

```python
f1880_name_count=''.join(f1880_name_set).lower()
f1940_name_count=''.join(f1940_name_set).lower()
f1990_name_count=''.join(f1990_name_set).lower()


f1880_name=Counter()
for i in f1880_name_count:
    f1880_name[i]+=1


f1940_name=Counter()
for i in f1940_name_count:
    f1940_name[i]+=1
    

f1990_name=Counter()
for i in f1990_name_count:
    f1990_name[i]+=1
```

```python
fre_count_1880=pd.DataFrame(data=f1880_name.values(),
                              index=f1880_name.keys(),
                              columns=['f1880_name'])

fre_count_1990=pd.DataFrame(data=f1990_name.values(),
                              index=f1990_name.keys(),
                              columns=['f1990_name'])

fre_count_1940=pd.DataFrame(data=f1940_name.values(),
                              index=f1940_name.keys(),
                              columns=['f1940_name'])

result_female = fre_count_1880.join([fre_count_1990,fre_count_1940])
result_female.index.name = 'Letter'

result_female=result_female.sort_values(by='Letter',ascending=True)
result_female
```

```python

result_female.head(30)

fig=plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(111)
x=result_female.index
y1=result_female.f1880_name
y2=result_female.f1940_name
y3=result_female.f1990_name

ax1.plot(x,y1,label='1880s')
ax1.plot(x,y2,label='1940s')
ax1.plot(x,y3,label='1990s')

ax1.set_ylabel('Frequecy')
ax1.set_xlabel('Letters')
ax1.set_title('Letter frequecy for Female in 1880s 1940s 1990s')
ax1.legend(title='Years')
plt.show()

```

```python
f1880_name_set=list(f1880_name_set)
last_letter_f1880=[i[-1] for i in f1880_name_set]
count_last_f1880=Counter()
for i in last_letter_f1880:
    count_last_f1880[i]+=1 
    
f1940_name_set=list(f1940_name_set)
last_letter_f1940=[i[-1] for i in f1940_name_set]
count_last_f1940=Counter()
for i in last_letter_f1940:
    count_last_f1940[i]+=1 
    
f1990_name_set=list(f1990_name_set)
last_letter_f1990=[i[-1] for i in f1990_name_set]
count_last_f1990=Counter()
for i in last_letter_f1990:
    count_last_f1990[i]+=1 
```

```python
fre_count_last_1880=pd.DataFrame(data=count_last_f1880.values(),
                              index=count_last_f1880.keys(),
                              columns=['frequence_last_letter_f1880'])
fre_count_last_1880.index.name = 'Last Letter'


fre_count_last_1990=pd.DataFrame(data=count_last_f1990.values(),
                              index=count_last_f1990.keys(),
                              columns=['frequence_last_letter_f1990'])
fre_count_last_1990.index.name = 'Last Letter'


fre_count_last_1940=pd.DataFrame(data=count_last_f1940.values(),
                              index=count_last_f1940.keys(),
                              columns=['frequence_last_letter_f1940'])
fre_count_last_1940.index.name = 'Last Letter'


count_last_total=count_last_f1990+count_last_f1940+count_last_f1880
fre_count_last_total=pd.DataFrame(data=count_last_total.values(),
                              index=count_last_total.keys(),
                              columns=['frequence_last_letter_total'])
fre_count_last_total.index.name = 'Last Letter'


fre_count_last_total.head(27)
```

```python
result_last_female = fre_count_last_total.join([fre_count_last_1880,fre_count_last_1940,fre_count_last_1990],
                                               how='left')
result_last_female
```

```python
result_last_female['frequence_last_letter_f1880']=result_last_female['frequence_last_letter_f1880'].fillna(0)
result_last_female['frequence_last_letter_f1940']=result_last_female['frequence_last_letter_f1940'].fillna(0)
result_last_female['frequence_last_letter_f1990']=result_last_female['frequence_last_letter_f1990'].fillna(0)
result_last_female=result_last_female.sort_values(by='Last Letter',ascending=True)
result_last_female
```

```python
fig=plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(111)
x=result_last_female.index
y1=result_last_female.frequence_last_letter_f1880
y2=result_last_female.frequence_last_letter_f1940
y3=result_last_female.frequence_last_letter_f1990
y4=result_last_female.frequence_last_letter_total
ax1.plot(x,y1,label='1880s')
ax1.plot(x,y2,label='1940s')
ax1.plot(x,y3,label='1990s')
ax1.plot(x,y4,label='Total')
ax1.set_ylabel('Frequecy')
ax1.set_xlabel('Last Letters')
ax1.set_title('Last Letter frequecy for Female in 1880s 1940s 1990s')
ax1.legend(title='Years')
plt.show()
```
