### Matrix Multiplication

```python
import numpy as np

M = np.array([[3,0],[0,-5]])

v = np.array([1,2])
M@v
```

Not to be confused with multiplication, which always does scalar multiplication, sometimes with
the help of broadcasting.  In
the case of `M` and `v`, the scalars in `v` will muliply the corresponding columns in `M`.

```python
M*v
```

## Eigenvectors and eigenvalues

```python
evals, evecs = np.linalg.eig(M)
evals, evecs
```

```python
def check_evals_evecs (evals,evecs):
    for idx in range(len(evals)):
        display(M@evecs[:,idx])
        display(evals[idx]*evecs[idx])
        print()


check_evals_evecs (evals,evecs)
```

### Checking without a loop

Matrix multiplication will apply `M` successively to each column of `evecs` without a loop.

```python
M@evecs
```

The rules of broadcasting will ensure that the scalars in `evals` will get multipleid by the corresponding
columns

```python
evals*evecs
```

```python
def check_evals_evecs (evals,evecs):
    display(M@evecs)
    display(evals*evecs)

check_evals_evecs (evals,evecs)
```
