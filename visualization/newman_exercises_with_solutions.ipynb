```python
from math import sqrt,sin,pi
from numpy import empty
from pylab import imshow,gray,show

wavelength = 5.0
k = 2*pi/wavelength
xi0 = 1.0
separation = 20.0      # Separation of centers in cm
side = 100.0           # Side of the square in cm
points = 500           # Number of grid points along each side
spacing = side/points  # Spacing of points in cm


# Calculate the positions of the centers of the circles
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2

# Make an array to store the heights
xi = empty([points,points],float)

# Calculate the values in the array
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = sqrt((x-x1)**2+(y-y1)**2)
        r2 = sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*sin(k*r1) + xi0*sin(k*r2)

# Make the plot
imshow(xi,origin="lower",extent=[0,side,0,side])
gray()
show()
```

#### A note on the effects of color normalization

In the unnormalized plot on the left, notice the complete disappearance of the dark concentric halos 
that are clearly visible around each source
in Newman's original plot in gray.  You can make the halos return in this visualization
by using the grayscale colormap `plt.cm.bone` (in which case you need to turn off color normalization as well)
or `plt.cm.magma` or `plt.cm.copper`. Alternatively, using a PowerNorm for color normalization helps bring them back as shown on the right. With other normalizations (such as Log normalization), the halos disappear.

Note that we are using the same data in both
pictures (and even the same color map!); only the way in which color represents wave amplitude differs.
This is something about how color and shape perception interact. 

If you look carefully at the halos you see they are something of an illusion.  They're discontinuous.
They are constructed by the eye by joining dark arcs 
with other dark arcs that lie across an interference ray, but are slightly misaligned.
Nevertheless the eye assembles such slightly misaligned
arcs into a single circle.  Exactly the same arcs exist in the picture 
on the left but the eye refuses to assemble them into halos because of the way the interference rays 
and their boundaries have been  colored (bright white).

The picture on the right looks more like what most lab wave interference experiments produce,
but it's not entirely clear which picture is better or truer.

## Millikan Photoelectric Effect Experiment:   Newman exercise 3.8 (p. 123)

The whole source of excitement about the photolectric effect was that it is governed by a
very simple equation, whose verification turned out to be very strong evidence that light consists of particles
(let's call them photons).

$$
\text{V} = \frac{h}{e} \nu - \Phi
$$

Read this as V (Voltage) = $\frac{h}{e}$ (Planck's Constant divided by the charge) times
$\nu$ (the frequency of the light) minus $\Phi$ (known as the work function of the surface).
So an electron gets kicked up from the surface by a photon, loses a little energy ($\Phi$)
because it takes some work to pry it loose, and then has energy (V) proportional to the
light's wavelength.

The experimental setup was such that the work function of the surface was constant throughout
the experiment.

Here is Millikan's data, measuring the voltage produced by single photons of light
at various frequencies. He won the Nobel Prize for these results, which confirmed
predictions made in Einstein's 1905 work on the photoelectric effect, for which Einstein also
won the Nobel Prize.

```python
import pandas as pd

e_charge = 1.602e-19
df = pd.read_csv("millikan.txt",header=None,delimiter= " ",names=("Frequency","Voltage"))
```

## Catalan numbers

Solution to Newman's Catalan Numbers problems: 2.7 (p. 46) and  2.13 (pp. 83,84). First print Catalan vals up to a billion, using two versions, the closed form formula see [the Catalan Numbers Wikipedia page](https://en.wikipedia.org/wiki/Catalan_number) and the simple intuitive
`next_catalan` function.

```python
from scipy.special import comb


def next_catalan (n, Cn):
    """
    Given n and Cn, compute C_{n+1}
    """
    return int((4*n +2)/(n + 2) * Cn)

##  Closed form expression
def catalan_closed_form (n):
    """
    Compute Cn given n using one of the closed form formulas. 
    
    Note this makes it clear that a Catalan number is always an integer.
    """
    #return int((1/(n+1))* comb(2*n,n))
    # Avoiding the need to cast as int
    return comb(2*n,n, exact=True)//(n+1)
```

```python
c,n = 1,0
c2,n2 = 1,1

while c < 1e9:
    print(f"{n:>2} {c:> 12,}")
    print(f"{n2:>2} {c2:> 12,}")
    print()
    c = next_catalan(n,c)
    c2 = catalan_closed_form(n2)
    n  += 1
    n2 += 1
```

Next, iterative and recursive functions for computing Catalan numbers, given $n$.

```python
def iterative_catalan (n):
    c = 1
    for i in range(n):
        c = ((4*i + 2)/(i + 2)) * c
    return int(c)

def recursive_catalan (n):
    if n == 0:
        return 1
    # Equivalent: harder to understand.
    #m=n-1
    #return int((4*m +2)/(m + 2) * recursive_catalan(m))
    return int((4*n - 2)/(n + 1) * recursive_catalan(n-1))
```

```python
for i in range(10):
    print(i, recursive_catalan (i))
    print(i, iterative_catalan(i))
    print()
        
```

## instability of quadratic formula:  Newman's syndrome

```python
import numpy as np

def q0 (a,b,c):
    """
    This just codes up the canonical version of the quadratic formula.
    """
    discriminant_rt = np.sqrt(b**2 - 4*a*c)
    return np.array([(-b + discriminant_rt)/(2*a), (-b-discriminant_rt)/(2*a)])

def q1 (a,b,c):
    """
    Second equally valid form of the quadratic formula 
    (multiply numerator and denominator by -b +/- s)
    """
    discriminant_rt = np.sqrt(b**2 - 4*a*c)
    return np.array([(2 * c)/(-b + discriminant_rt), (2 * c)/(-b - discriminant_rt)])

# .001x^2 + 1,000x + .001
a,b,c = 1e-3, 1e3, 1e-3

# discriminant sum in numerator, b is positive, so smaller mag solution comes first
sm_mg0,lg_mg0 = q0(a,b,c)
# discriminant sum in denominator, so smaller mag solution comes second
lg_mg1,sm_mg1 = q1(a,b,c)
print(f"Q1 {sm_mg0:.16f} {lg_mg0:.16f} Q2: {sm_mg1:.16f} {lg_mg1:.16f}" )
```

Ok, so two completely equivalent versions of the quadratic formula produce different solutions
for the equation

$$
.001x^2 + 1,000x + .001 = 0.
$$

That's already pretty weird.  Let's check the quality of the solutions.

```python
#  Checking quality of solns
f = lambda x : a*(x**2) + (b * x) + c
print(f"Q1 {f(sm_mg0):.16f} {f(lg_mg0):.16f} Q2: {f(sm_mg1):.16f} {f(lg_mg1):.16f} " )
```

Not a very clear picture.  One solution looks to be perfect.  One appears terrible.

Let's try to get the big picture by eyballing a plot:

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(8,8))
x = np.arange(-1_110_000,50_000,50)
f = lambda x : a*(x**2) + (b * x) + c
y = f(x)

ax.axhline()
ax.axvline()
_ = ax.plot(x,y)
```

Based on the plot we see the two solitions are: one very close to -1_000_000
and one very close to 0.

The plot makes one wonder if 0 is a solution, but it isn't quite:

```python
# Is 0 a soln?  No.
print(f"{f(0):.16f}")
```

In fact, both versions of quabdratic formula seem to find better approximations of the solution close to 0.

So we have 4 solutions, two from each form of the quadratic formula; we'll call them the Q0 solutions and the Q1 solutions.  In each case, we'll call the one close to -1,000,0000 the **large magnitude solution** (lg_mg) and the one close to 0 the **small magnitude solution** (sm_mg). We'll call the corresponding correct solutions the lm-root and sm-root. The magnitude of a number is its absolute value.  

By plugging the lg_mg-solutions into $f$, we can see that the lg_mg-solution from Q0 is clearly closer to the the lm root than the lg_mg-solution from Q1:

```python
print(f"Q0 {f(lg_mg0):.16f} Q1:  {f(lg_mg1):.16f} %-Error: {np.abs(lg_mg1-lg_mg0)/np.abs(lg_mg0):.4%} " )
```

The third number measures the error-percentage in the Q1-solution by using the Q0 solution as the true value.  That may seem like a small number but on fact it's many orders of magnitude greater than the error one should
expect from a floating point calculation done out of the blue:

```python
print(f"{1e-16:.16%}")  
```

That's unacceptable error.

Similarly. the sm_mg-solution from Q1 is clearly closer to the the sm-root than the sm_mg-solution from Q0:

```python
print(f"Q0 {f(sm_mg0):.16f} Q1:  {f(sm_mg1):.16f} %-Error: {np.abs(sm_mg0-sm_mg1)/np.abs(sm_mg1):.4%} " )
```

The Q0-solution doesnt look so bad because `f(sm_mg0)` is pretty close to 0, but in fact, its error-percentage
is about the same as the error percentage we saw with large magnitude solutions.

Bottom line:  Both forms of the quadratic formula have problems.  Each is completely correct mathematically speaking and yet each computes a solution with a completely unacceptable error percentage. Yet these are just numerical calculations.  Number-crunching is what computers excel at.  What's going on?

The issue here is a well-known problem that arises in floating point arithmetic for
subtractions involving large numbers that are close in value.  (Using completely nonstandard
terminology, I'm going to call this problem **Newman's Syndrome**, because I learned about it on
pp. 131-133 of *Computational Physics*.)

By subtraction here I mean computational subtraction: any summing operation $a + b$ whose result $r$ is such that:
$$
\mid\, r\!\mid < \,\mid\!a \!\mid \,\text{ or }\, \,\mid\!r\!\mid <\, \mid\!b\,\mid
$$ 
Essentially, that happens when $a$ and $b$ differ in sign.  So all the following involve subtraction:
1  - 2, 2 - 1, 1 + (-2), -1 + 2.


Now consider the canonical version of the Quadratic fornula (Q0) and the given equation:

$$
x =  \frac{-b \pm \sqrt{b^{2} - 4ac}}{2a}
$$

$$
.001x^{2} + 1,000x + .001  = 0
$$

The values for $a$, $b$, and $c$ have been set up so that $b^{2} - 4ac$ is very close to $b^{2}$; therefore,
since $b$ is positive, $-b + \sqrt{b^{2} - 4ac}$ will exhibit Newman's Syndrome.  It will be a subtraction
involving large numbers that are close together.  And indeed the solution that has
the unacceptable error is the solution that uses $-b + \sqrt{b^{2} - 4ac}$.

So what is the problem with such subtractions?  Consider Newman's nice example

$$
x = 1,\quad y = 1 + 10^{-14}\sqrt{2}
$$

It's clear that $(y -x)$ introduces Newman's syndrome and it's also clear that:

$$
10^{14}(y-x) = \sqrt{2}
$$

What does Python tell us?

```python
x, y = 1, 1 + (1e-14)*np.sqrt(2)
z = 1e14*(y-1)
print(f"True: {np.sqrt(2)} Computed: {z} %-Error {np.abs(np.sqrt(2)-z)/np.sqrt(2):.4%}")
```

The error percentage in computing $z$ is huge, nearly half a percent.

Explanation: the error percentage in computing $y = 1+ (1e^{-14})\sqrt{2}$ is just normal floating point error,
a tiny percentage of the true value of $1 + (1e^{-14})\sqrt{2}$.  But when we do the subtraction
to compute $y - 1$, that same error is now transmitted to the computation of a much smaller value, meaning 
the error percentage just skyrocketed. 

Takeaway:  Newman's Syndrome occurs when subtracting numbers that are really close in magnitude (b-b'); that's when small errors in computing b and b' can become large in (b-b'), where large means large relative to the
size of (b-b'). So the unacceotable errors arose not because of
any problem with the Quadratic Formula (in either form), but because
the particular values for $a$, $b$, and $c$ we used created a problematic
floating point calculation when plugged into the formula.  The description usually used for the computational problem is that the quadratic formula is **numerically unstable** when $b$ is large relative to $a$ and $b$.

Fortunately there is a simple fix.  You should think about what the fix might be before moving onto the discussion below, which codes up the fix.

#### Fixing things

We fix things by avoiding subtraction.  Consider the following rule:

>**Rule A**: When b is positive we avoid Newman's Syndrome with the `-b  - discriminant`
form. When b is negative we avoid it with the `-b + discriminant` form.

Neither case will allow a small error to be turned into a big error because the
terms being combined agree in sign. There is no "subtraction",
so no Newman's Syndrome.

So for concreteness, let's say we use Q0, and apply Rule A. 
That gives us one solution that is safe from Newman's Syndrome.

But what do we do about the other solution? Well, essentially we get that from
Q1.  Let's take the case where `b` is positive.  Then the safe solution is the
one that uses `-b  - discriminant` form of the discriminant sum.  Since the
discriminant sum occurs in the numerator of Q0 and involves a subtraction,
it will be the larger in magnitude solution.  If we use that same form of the discriminant sum in Q2,
we get the smaller in magnitude solution, since the discriminant sum occurs in the denominator.
Parallel reasoning applies to the case where `b` is negative.

This leads directly to the new code below.


#### The new code

```python
def quadratic_roots(a,b,c):
    # solution orders: 1st solns use discriminant sum -b+d, second use -b-d
    q0sol1,q0sol2 = q0(a,b,c)
    q1sol1,q1sol2 = q1(a,b,c)
    if b > 0:
        # we want the -b-d solution forms
        return q0sol2,q1sol2
    else:
        # we want the -b+d solution forms
        return q0sol1,q1sol1
```

```python
# .001x^2 + 1,000x + .001
a,b,c = 1e-3, 1e3, 1e-3
sol1, sol2 = quadratic_roots(a,b,c)
print(f"Q    {sol1:.16f} {sol2:.16f}" )
print(f"f(Q) {f(sol1):.16f} {f(sol2):.16f}" )
```

```python
# .001x^2 + 1,000x + .001
a,b,c = 1e-3, -1e3, 1e-3
sol1, sol2 = quadratic_roots(a,b,c)
print(f"Q    {sol1:.16f} {sol2:.16f}" )
print(f"f(Q) {f(sol1):.16f} {f(sol2):.16f}" )
```

#### Addendum:  A small efficiency improvement.  

From [the Wikipeida page on the quadratic formula.](https://en.wikipedia.org/wiki/Quadratic_equation)

```python
# smaller-in-magnitude root can be computed as (c/a)/R where R is the bigger-in-magnitude root

bg_mg_sol = np.abs(q0(a,b,c)).max()
s5 = (c/a)/bg_mg_sol
print(f"{s5:.16f} {f(s5):.16f}")
```

Proof

Suppose $\frac{-b + \sqrt{d}}{2a}$ is the larger magnitude solution ($lm$)
computed by Q0.  The
first thing to observe is that the $lm$-solution can never be the solution
suffering from Newman's Syndrome, since the Q0-solution that uses
subtraction in the discriminant sum will always be lower in magnitude than
the one that does not (proof left as an exercise).

$$
lm = \frac{-b + \sqrt{d}}{2a}
$$

$$
\frac{c/a}{lm} = \frac{c}{a}\cdot \frac{2a}{-b + \sqrt{d}} = \frac{2c}{-b + \sqrt{d}} 
$$

where $\frac{2c}{-b + \sqrt{d}}$ is of course the non-mm solution
computed by Q1(a,b,c). The reasoning when $\frac{-b - \sqrt{d}}{2a}$
is the larger magnitude ($lm$) solution is quite similar and is left as an exercise.


```python
def quadratic_roots2(a,b,c):
    bg_mg_sol = np.abs(q0(a,-b,c)).max()
    return bg_mg_sol,(c/a)/bg_mg_sol
```

```python
bg_mg_sol2, sm_mg_sol2 = quadratic_roots2(a,b,c)
print(f"Q {sm_mg_sol2:.16f} {bg_mg_sol2:.16f}" )
print(f"Q {f(sm_mg_sol2):.16f} {f(bg_mg_sol2):.16f}" )
```
