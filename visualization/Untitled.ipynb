```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.arange(1,10,.01)
y = np.e/x

fig,ax = plt.subplots(1,1,figsize=(10,8))
ax.plot(x,y,label="$xy = e$")
#f=10
f = -1
temp = f"x/{f}"
# Latex needs string with actual {...} --- cant use f-string
ax.plot(x,np.e**(x/f),label="$e^{"+temp+"}$")
ax.legend()
```
