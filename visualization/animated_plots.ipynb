````python
# A sasmple function that draws each frame of the animation
# i is a required parameter: frame index.

def animate(i):
    global x, y, ax
    pt = randint(1,9) # grab a random integer to be the next y-value in the animation
    x.append(i)
    y.append(pt)

    ax.clear()
    ax.plot(x, y)
    # Need to reset bds each time after ax.clear()
    ax.set_xlim([0,20])
    ax.set_ylim([0,10])
````

"Our `animate()` function is defined, now we need to call the animation. Matplotlib's `FuncAnimation` class can accept several input arguments. At a minimum, we need to pass in the figure object fig, and our animation function that draws the plot animate to the FuncAnimation class. We'll also add a `frames=` keyword argument that sets many times the plot is re-drawn meaning how many times the animation function is called. `interval=500` specifies the time between frames (time between animate() function calls) in milliseconds. `interval=500` means 500 milliseconds between each frame, which is half a second. `repeat=False` means that after all the frames are drawn, the animation will not repeat. Note how the `plt.show()` line is called after the FuncAnimation line.

````python
from random import randint

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import numpy as np

fig, ax = plt.subplots()

x = []
y = []

t = np.linspace(0,2*np.pi)
x = np.sin(t)
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])
# run the animation interval unit is ms so 500 is quite slow
ani = FuncAnimation(fig, animate, frames=len(t),interval=250,repeat=False)

# alternative 
#ani = FuncAnimation(fig, animate, frames=20, interval=500, repeat=False)
HTML(ani.to_html5_video())
````

````python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t),repeat=False)

plt.show()
````
