### Geometric reframing


Let's reframe this geometrically in a way that will
help us understand how useful equation (1)
is.  Let L' be a line through the origin parallel to L (shown
as a green line in the picture).  Then since $\vec{w_{}}$ is perpendicular to L',
the dot product of $\vec{w_{}}$ with any point on L' is 0. Any point P on
L lies directly above a point P' on L' such that
$P = P' + (0,\, -b/w_{2})$.  For example, if we consider the intercept of L $(0,\, -b/w_{2})$, 
then P' is (0,0).  Using this way of taking points on P apart, consider 
$\vec{w}\cdot P$:

$$
\begin{array}[t]{llcll}
(3) & \vec{w_{}} \cdot P &=& \vec{w_{}} \cdot P' & + \,\vec{w_{}} \cdot (0,\,-b/w_{2})\\
    &      &=& 0 & + \,w1\cdot 0 +\, w_{2}\cdot -b/w_{2} \\
    &      &=& -b.
\end{array}
$$


Now to see how the classifier classifies, let's rewrite (1):

$$
\begin{array}[t]{llcl}
(1') & \vec{w_{}} \cdot \vec{x_{}} + b & =& 0
\end{array}
$$

and let's define the decision function $d$:

$$
\begin{array}[t]{llcl}
(4) &  d(X) =  \vec{w_{}} \cdot \vec{x_{}} + b 
\end{array}
$$

Consider applying $d$ to any point $\vec{x}$ in the data set. If $\vec{x}$ is on
the separator line L, $d(\vec{x}) = 0$, by (1').   If $\vec{x}$ is a positive point, then it
lies at some positive y-displacement $\delta$ above L and hence at displacement $-b/w2 + \delta$ above
L'.  Hence we can compute $w \cdot \vec{x}$
along the lines of (3):

$$
\begin{array}[t]{lcll}
w \cdot \vec{x} &=& w \cdot (P' & + \,(0,\,-b/w_2) +\, (0, \delta)\,)\\
          &=& w \cdot P' & + \, w \cdot (0,\,-b/w_2) +\, w \cdot (0, \delta)\\
          &=& 0 & + \,(w_1\cdot 0) + \,(w_{2}\cdot -b/w_2) + \,(w_1\cdot 0) \,+ (w_2\cdot \delta)\\
          &=&  & -b + w_2\cdot \delta\\
\end{array}
$$


Hence

$$
\begin{array}[t]{lclll}
d(\vec{x}) &=& \vec{w} \cdot \vec{x}  + b &=& w_2\cdot \delta           
\end{array}
$$

Without loss of generality we may chose $w_2$ to be positive, so that $d$ always
has a positive value for positive class
points,  and a negative value for negative class points (where $\delta$ will be negative).

Hence the class our classifier assigns is:

$$
\begin{array}[t]{lclll}
(5) & \text{sgn}(d(\vec{x})) & = & \text{sgn}(\vec{w} \cdot \vec{x} + b)
\end{array}
$$

By definition, the sign function returns 1 for positive
values and -1 for negative. In  keeping with (5), then SVM formulations define $y_{i}$,
the class of $x_{i}$,  to be 1 for positive class instances and -1 for negative class instances.
